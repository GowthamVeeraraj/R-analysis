#Feature Engineering

Batting Average
Onbase Percentage
Slugging Percentage

```r
df1$BA = df1$H/df1$AB
```

```r
tail(df1$BA)
```

```r
df1$OBP = (df1$H+df1$BB+df1$HBP)/(df1$AB+df1$BB+df1$HBP+df1$SF)
```

```r
tail(df1$OBP)
```

```r
df1$X1B = df1$H-df1$X2B-df1$X3B-df1$HR
```

```r
df1$SLG = ((1 * df1$X1B)+(2 * df1$X2B)+(3 * df1$X3B)+(4 * df1$HR))/df1$AB 
```
